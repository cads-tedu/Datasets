<b>19)</b> lkelerin kazandklar madalyalarn ortanca deerinden fazla madalya kazanan lkelerin ortalama, ortanca, minimum ve maximum gayri safi yurtii hasla miktarlar nedir?
    
<b>pucu:</b>
1. Bir nceki soruda oluturulan Dataframe'e gdp ismi verelim.
2. gdp veri setini madalya saylar medyan madalya saysndan bykleri kapsayacak ekilde filtreleyelim. gdp.Madalya.median()
3. .agg() metodu kullanarak ilgili istatistikleri elde edelim.
</div>

```python
#Buraya Kodlayn
```

```python
gdp = pd.DataFrame(olimpiyatlar['Country'].value_counts()).reset_index().rename(columns = {'index':'Country', 'Country':'Madalya'}).merge(kunye, how = 'right').sort_values('Madalya',ascending = False)

gdp[gdp['Madalya'] > gdp.Madalya.median()]['GDP'].agg({'Maks_GDP':'max',
                                                      'Min_GDP':'min',
                                                      'Medyan_GDP':'median',
                                                      'Ortalama_GDP':'mean'})

#veya

gdp[gdp['Madalya'] > gdp.Madalya.median()].agg(Maks_GDP = ('GDP', max),
                                              Min_GDP = ('GDP', min),
                                              Medyan_GDP = ('GDP', np.median),
                                              Ortalama_GDP = ('GDP', np.mean))
```

3. .agg() fonksiyonu kullanarak Madalya_Says adnda bir kolon oluturulur 'Medal' kolonuna 'count' fonksiyonu Python'da varsaylan olarak bulunmad iin fonksiyon lambda fonksiyonuyla oluturulup bu kolona uygulanr. lambda x: x.count()

```python
olimpiyatlar[olimpiyatlar['Country'] == 'Turkey'].groupby(['City', 'Sport']).agg(Madalya_Says = ('Medal', lambda x: x.count()))
#Bonus
olimpiyatlar[olimpiyatlar['Country'] == 'Turkey'].pivot_table(values = 'Medal', index='City', columns='Sport', aggfunc=lambda x: x.count())
```
